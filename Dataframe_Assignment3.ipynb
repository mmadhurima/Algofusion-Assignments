```python
df1=pd.read_csv("Dataset_assignment.csv",index_col="Subject_Name")
df1
```

```python
def category(x):
    if x>90:
        return "Excellent"
    if x>80 and x<90 :
        return "very good"
    if x>70 and x<80:
        return "good"
    if x>60 and x<70:
        return "better"
    return "average"
df1=df1["Marks scored"].apply(category)
```

```python
df1
```

```python
df1.describe()
```
